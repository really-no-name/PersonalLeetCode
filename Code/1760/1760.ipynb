# 袋子里最少数目的球

https://leetcode.cn/problems/minimum-limit-of-balls-in-a-bag?envType=daily-question&envId=2025-02-12

给你一个整数数组 nums ，其中 nums[i] 表示第 i 个袋子里球的数目。同时给你一个整数 maxOperations 。

你可以进行如下操作至多 maxOperations 次：

选择任意一个袋子，并将袋子里的球分到 2 个新的袋子中，每个袋子里都有 正整数 个球。
比方说，一个袋子里有 5 个球，你可以把它们分到两个新袋子里，分别有 1 个和 4 个球，或者分别有 2 个和 3 个球。
你的开销是单个袋子里球数目的 最大值 ，你想要 最小化 开销。

请你返回进行上述操作后的最小开销。

 

示例 1：

> 输入：nums = [9], maxOperations = 2
> 
> 输出：3
> 
> 解释：
> - 将装有 9 个球的袋子分成装有 6 个和 3 个球的袋子。[9] -> [6,3] 。
> - 将装有 6 个球的袋子分成装有 3 个和 3 个球的袋子。[6,3] -> [3,3,3] 。
> 装有最多球的袋子里装有 3 个球，所以开销为 3 并返回 3 。

示例 2：

> 输入：nums = [2,4,8,2], maxOperations = 4
> 
> 输出：2
> 
> 解释：
> - 将装有 8 个球的袋子分成装有 4 个和 4 个球的袋子。[2,4,8,2] -> [2,4,4,4,2] 。
> - 将装有 4 个球的袋子分成装有 2 个和 2 个球的袋子。[2,4,4,4,2] -> [2,2,2,4,4,2] 。
> - 将装有 4 个球的袋子分成装有 2 个和 2 个球的袋子。[2,2,2,4,4,2] -> [2,2,2,2,2,4,2] 。
> - 将装有 4 个球的袋子分成装有 2 个和 2 个球的袋子。[2,2,2,2,2,4,2] -> [2,2,2,2,2,2,2,2] 。
> 装有最多球的袋子里装有 2 个球，所以开销为 2 并返回 2 。

示例 3：

> 输入：nums = [7,17], maxOperations = 2
> 
> 输出：7
 

提示：

- `1 <= nums.length <= 105`
- `1 <= maxOperations`, `nums[i] <= 109`

## 分析

### 首先排除两种特殊情况，即所有袋内的球均为1，或操作数大于总球数，直接返回1

### 进行二分查找

每个袋内的球的数量范围是 1 - 初始袋内最多的球数 --> `left, right = 1, max(nums)`

`mid` 是当前的中间值，表示我们尝试的最大球数。

`operation` 计算将每个袋子中的球数减少到 mid 或以下所需的操作次数。


```
for num in nums:
    operations += (num - 1) // mid
```
这个 `for` 循环的作用是计算将每个袋子中的球数减少到不超过 `mid` 所需的操作次数。具体来说，`(num - 1) // mid` 是计算将 `num` 个球分成若干个袋子，每个袋子中的球数不超过 `mid` 所需的操作次数的关键表达式。

下面我们从数学角度详细解释这个表达式的含义。

1. **问题背景**
    - 我们有一个袋子，里面有 `num` 个球。
    - 我们需要将这些球分成若干个袋子，每个袋子中的球数不超过 `mid`。
    - 每次操作可以将一个袋子分成两个袋子，且每个新袋子中的球数必须为正整数。
    我们的目标是计算将 `num` 个球分成若干个袋子，每个袋子中的球数不超过 `mid`，需要多少次操作。

2. **操作次数的计算**

    假设我们有一个袋子，里面有 `num` 个球，我们需要将其分成若干个袋子，每个袋子中的球数不超过 `mid`。每次操作可以将一个袋子分成两个袋子。
   
    分析：
        1. 如果 `num <= mid`，则不需要任何操作，因为袋子中的球数已经满足条件。
        2. 如果 `num > mid`，则需要将袋子分成多个袋子，每个袋子中的球数不超过 `mid`。

    操作次数的推导：
        - 每次操作可以将一个袋子分成两个袋子。因此，每增加一次操作，袋子的总数会增加 1。
        - 假设最终有 `k` 个袋子，每个袋子中的球数不超过 `mid`，那么需要的操作次数为 `k - 1`（因为每次操作增加一个袋子）。

    计算 `k`：
        - 我们需要将 `num` 个球分成 `k` 个袋子，每个袋子中的球数不超过 `mid`。
        - 最小的 `k` 满足：`k * mid >= num`。
        - 因此，`k` 的最小值为 `ceil(num / mid)`，即 `(num + mid - 1) // mid`。

    操作次数：
        - 操作次数为 `k - 1`，即 `(num + mid - 1) // mid - 1`。
        - 化简后，操作次数可以表示为 `(num - 1) // mid`。
3. **公式推导**
    
   我们可以通过数学推导来证明 `(num - 1) // mid` 是正确的操作次数。

    目标：
        
        计算将 `num` 个球分成若干个袋子，每个袋子中的球数不超过 `mid`，所需的最少操作次数。

    推导过程：
   
        > 1. 假设最终有 `k` 个袋子，每个袋子中的球数不超过 `mid`。
        > 
        > 2. 因为每次操作增加一个袋子，所以操作次数为 `k - 1`。
        > 
        > 3. 最小的 `k` 满足：`k * mid >= num`。
        >    - 这是因为每个袋子最多有 `mid` 个球，所以 `k` 个袋子最多可以容纳 `k * mid` 个球。
        > 
        > 4. 解不等式 `k * mid >= num`，得到 `k >= ceil(num / mid)`。
        > 
        > 5. 因为 `k` 必须是整数，所以 `k = ceil(num / mid)`。
        > 
        > 6. 操作次数为 `k - 1 = ceil(num / mid) - 1`。
        > 
        > 7. 在 Python 中，`ceil(a / b)` 可以表示为 `(a + b - 1) // b`。
        >    - 因此，`ceil(num / mid) = (num + mid - 1) // mid`。
        > 
        > 8. 操作次数为 `(num + mid - 1) // mid - 1`。
        > 
        > 9. 化简：
        >    \[
        >    (num + mid - 1) // mid - 1 = (num - 1) // mid
        >    \]
        >    - 这是因为 `(num + mid - 1) // mid - 1` 等价于 `(num - 1) // mid`。



## 511 ms, 29.02 mb

````python
from typing import List


class Solution:
    def minimumSize(self, nums: List[int], maxOperations: int) -> int:
        if max(nums) == 1:
            return 1
        if maxOperations >= sum(nums):
            return 1
        
        left, right = 1, max(nums)  # 二分查找的左右边界
        while left < right:
            mid = (left + right) // 2  # 当前中间值， 尝试的最大球数
            operation = 0
            
            for num in nums:
                operation += (num - 1) // mid
                
            if operation <= maxOperations:
                right = mid
            else:
                left = mid + 1
        return left
    
    
if __name__ == '__main__':
    sol = Solution()
    print(sol.minimumSize([2,4,8,2], 4))
    print(sol.minimumSize([9], 200))
````
